```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
data = [
    "256 workers; 17m17.239s",
    "128 workers; 29m37.127s",
    "64 workers; 48m30.026s",
    "32 workers; 104m35.518s",
    "16 workers; 205m12.042s",
    "8 workers; 357m6.920s",
    "4 workers;  661m0.847s",
    "2 workers; 1280m5.768s",
    "1 worker; 2547m51.077s"
]

num_workers = []
times = []

for item in data:
    parts = item.split('; ')
    num_workers.append(int(parts[0].split()[0]))
    time_parts = parts[1].split('m')
    minutes = int(time_parts[0])
    seconds = float(time_parts[1][:-1])
    total_minutes = minutes + seconds / 60
    times.append(total_minutes)

print(num_workers)
print(times)
```

```python
plt.plot(num_workers, times)
plt.scatter(num_workers, times)
plt.title("Iris Segmentation on Casia Iris Lamp Data")
plt.xlabel("Number of Workers 1 to 256")
plt.ylabel("Time in Minutes")
plt.show()
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']

plt.figure(figsize=(12, 6))

for worker, time, color in zip(num_workers, times, colors):
    plt.scatter(worker, time, color=color, label=f"{worker}")

plt.plot(num_workers, times, color='gray', alpha=0.5)  # Plot a gray line for reference

plt.title("Iris Segmentation on Casia Iris Lamp Data", fontsize=16)
plt.xlabel("Number of Workers", fontsize=11)
plt.ylabel("Time in Minutes", fontsize=11)
plt.legend(title='Workers', loc='right', bbox_to_anchor=(1, 0.5))  # Add a legend
plt.show()
```
